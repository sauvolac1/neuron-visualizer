# FB Visualization Example

Generates an interactive 3D visualization of **FB** neurons (fan-shaped body tangential neurons).

This is a large dataset (~600 neurons, 166 types) with custom continuous color modes
derived from valence scores and sensory modality input strengths.

**Expect this to take ~20 minutes** on first run (synapse fetching is the bottleneck).
Subsequent runs reuse the saved synapse `.json` and are much faster.

```python
import sys, os
sys.path.insert(0, os.path.join(os.path.dirname(os.getcwd()), '..', 'Core_Code'))
from generate_visualization import generate_visualization

# --- neuPrint credentials ---

NEUPRINT_TOKEN = 'YOUR_TOKEN_HERE'
# Paste your token from neuprint-cns.janelia.org → Account → Auth Token

NEUPRINT_SERVER = 'neuprint-cns.janelia.org'

NEUPRINT_DATASET = 'cns'

# --- Configuration ---

PATTERN = '^FB.*'

CONTINUOUS_CSVS = [
    'Custom_Colors/fb_type_scores.csv',
    'Custom_Colors/fbt_sensory_modality_scores_unsigned.csv',
]
# fb_type_scores.csv: valence score per type (divergent → RdBu colormap)
# fbt_sensory_modality_scores_unsigned.csv: olfactory, gustatory, thermo, mechano,
#   hygro, visual input strengths (each becomes a separate color mode)

USE_MESHES = True

MESH_FACES = 'auto'

MAX_FILE_MB = 500

OUTPUT_DIR = '.'

# --- Generate ---

generate_visualization(
    PATTERN,
    continuous_csvs=CONTINUOUS_CSVS,
    use_meshes=USE_MESHES,
    mesh_faces=MESH_FACES,
    max_file_mb=MAX_FILE_MB,
    output_dir=OUTPUT_DIR,
    server=NEUPRINT_SERVER,
    dataset=NEUPRINT_DATASET,
    token=NEUPRINT_TOKEN,
    skip_synapses=True,
    auto_open=True,
)
```

```python
import re
from pathlib import Path
from add_synapses import add_synapses_to_html
from generate_visualization import get_client

get_client(server=NEUPRINT_SERVER, dataset=NEUPRINT_DATASET, token=NEUPRINT_TOKEN)

safe_name = re.sub(r'[\^$.*+?\[\](){}|\\]', '', PATTERN)
html_path = Path('.') / f'{safe_name}_visualization.html'
add_synapses_to_html(html_path)
```
